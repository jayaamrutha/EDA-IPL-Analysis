```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ipl_df = pd.read_csv("IPL Ball-by-Ball 2008-2020 (1).csv")
ipl_df
```

```python
ipl_df.describe()
```

```python
# to know the data types
ipl_df.info()
```

```python
ipl_df.isnull().sum()
```

```python
# 1 a Creating a DataFrame for batsman only
# 1 b Adding Average and Strikerate to batsman line up
df_strikerate = ipl_df.groupby(['batsman']).agg({'ball':'count','batsman_runs':'mean'}).sort_values(by='batsman_runs',ascending=False)
df_strikerate.rename(columns ={'batsman_runs' : 'strike_rate'}, inplace=True)

df_runs_per_inning = ipl_df.groupby(['batsman','id']).agg({'batsman_runs':'sum'})
df_total_runs = df_runs_per_inning.groupby(['batsman']).agg({'sum' ,'mean','count'})
df_total_runs.rename(columns ={'sum' : 'batsman_run','count' : 'match_count','mean' :'average_score'}, inplace=True)
df_total_runs.columns = df_total_runs.columns.droplevel()

df_sixes = ipl_df[['batsman','batsman_runs']][ipl_df.batsman_runs==6].groupby(['batsman']).agg({'batsman_runs':'count'})
df_four = ipl_df[['batsman','batsman_runs']][ipl_df.batsman_runs==4].groupby(['batsman']).agg({'batsman_runs':'count'})
df_batsman_stats = pd.merge(pd.merge(pd.merge(df_strikerate,df_total_runs, left_index=True, right_index=True),
                                    df_sixes, left_index=True, right_index=True),df_four, left_index=True, right_index=True)

df_batsman_stats.rename(columns={'batsman_runs_x' :'six','batsman_runs_y':'four'},inplace=True)
```

```python
df_batsman_stats.rename(columns = {'ball' : 'ball', 'strike rate':'strike_rate','batsman run' : 'batsman_run',
                                  'match count' : 'innings_count','average score' : 'average_score' ,'batsman_runs_x' :'six',
                                  'batsman_runs_y':'four'},inplace=True)
df_batsman_stats['strike_rate'] = df_batsman_stats['strike_rate']*100
df_batsman_stats.sort_values(by='batsman_run',ascending=False,inplace=True)
df_batsman_stats.reset_index(inplace=True)
df_batsman_stats
```

```python
# 1 c 
# creating a df for max strikereate for balls greater than 500
df_balls = df_batsman_stats[df_batsman_stats['ball'] > 500].sort_values(by = 'strike_rate',ascending = False)
df_balls
```

```python
# creating a df for max strikereate for matches greater than 50
df_matchcount =  df_balls[df_balls['match_count'] > 50].sort_values(by = 'strike_rate',ascending = False)
df_matchcount
```

```python
 #creating a df for max strikereate for average score greater than 25
df_top = df_matchcount[df_matchcount['average_score'] > 25].sort_values(by = 'strike_rate',ascending = False)
df_top
```

```python
# 1 c Finding the top 6 batsman
df_TopBatsman = df_top.head(6)
df_TopBatsman
```

```python
# 1 d creating a list for top 6 players in batting
list_of_top_batsman = list(df_TopBatsman['batsman'])
list_of_top_batsman
```

```python
# 2 a creating a df for bowlers data only

bowlers_df=ipl_df.groupby('bowler').sum().reset_index()
bowl_name=ipl_df['bowler'].value_counts().reset_index()
bowlers_df=bowlers_df.merge(bowl_name,left_on='bowler',right_on='index',how='left')
bowlers_df=bowlers_df[['bowler_x','total_runs','bowler_y','inning']]
bowlers_df.rename({'bowler_x':'bowler','total_runs':'runs_given','bowler_y':'balls','inning':'match_count'},axis=1,inplace=True)
bowlers_df['overs']=(bowlers_df['balls']//6)
bowlers_df
```

```python
# 2 b Getting the economy rate for bowlers
dismissal_kinds = ["bowled", "caught", "lbw", "stumped", "caught and bowled", "hit wicket"]  
ct=ipl_df[ipl_df["dismissal_kind"].isin(dismissal_kinds)]
ct=ct['bowler'].value_counts().reset_index()
bowlers_df=bowlers_df.merge(ct,left_on='bowler',right_on='index',how='left').dropna()
bowlers_df=bowlers_df[['bowler_x','runs_given','overs','bowler_y','match_count']]
bowlers_df.rename({'bowler_x':'bowler','bowler_y':'wickets'},axis=1,inplace=True)
bowlers_df['avg_economy']=(bowlers_df['runs_given']/bowlers_df['overs'])
bowlers_df
```

```python
# 2 c bowlrers min economy rate for 500 matches
df_bowlers = bowlers_df[bowlers_df['match_count'] > 50].sort_values(by = 'avg_economy',ascending = True)
df_bowlers
```

```python
# 2 c Bowlers economy rate for over greater than 30
df_Wickets = df_bowlers[df_bowlers['wickets'] > 50].sort_values(by = 'avg_economy',ascending = True)
df_Wickets
```

```python
# 2 c bowler with min economy for overs greater than 200 
df_Overs = df_Wickets[df_Wickets['overs'] > 200].sort_values(by = 'avg_economy',ascending = True)
df_Overs
```

```python
# 2 c Top 6 bowlers 
df_TopBowlers = df_Overs.head(6)
df_TopBowlers
```

```python
# 2 d Identifying top 5 bowlers and making a python list
df_Bowlingteam = df_TopBowlers.head(5)
list_of_top_bowlers = list(df_Bowlingteam['bowler'])
list_of_top_bowlers
```

```python
# 2 e Making the final players lineup team 
df_finalplayers = list_of_top_batsman + list_of_top_bowlers
print(df_finalplayers)
```
